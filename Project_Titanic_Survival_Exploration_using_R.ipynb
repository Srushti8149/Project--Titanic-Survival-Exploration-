```r
# Create the input vectors.
colors = c("green","orange")
pclass <- c(1,2,3)
type <- c("died", "survived")

# STACKED BAR CHART WITH COLORS AND LEGEND
counts = table(df$Survived, df$Pclass)

# CONSTRUCT BARCHART
barplot(counts, 
        main = "Titanic Survival Rates by Passenger Class",
        xlab = "Passenger Class", 
        col = terrain.colors(2))

# ADD LEGEND
legend("topleft", 
       inset = .03,
       legend = c("died", "survived"), 
       fill = terrain.colors(2), 
       horiz = TRUE)            

# This stacked bar plot gives us Survival Rates by its passanger class. It gives the count of died & survived persons by its passanger class.
```

```r
counts = table(df$Survived, df$Sex)

barplot(counts, 
        main = "Titanic Survival Counts by Passenger Sex",
        xlab = "Passenger Sex", 
        ylab = "Count",
        col = topo.colors(2),  
        beside = TRUE)

legend("topleft", 
       inset =.03,              # The optional inset argument specifies how far the legend is inset from the plot margins. If a single value is given, it is used for both margins
       legend = c("died", "survived"), 
       fill = topo.colors(2), 
       horiz = TRUE)

# This barplot gives us survival count by passanger sex. it gives the dird and survived passangers count by their sex.
```

```r
ggplot(df, aes(x=Fare, y=Parch)) + geom_point()            #scatterplot     The function geom_point() adds a layer of points to plot, which creates a scatterplot.
```

```r
g <- ggplot(df, aes(x=Fare, y=Parch)) + geom_point() + geom_smooth(method="lm")  # set se=FALSE to turnoff confidence bands
plot(g)     

# By default, this is set to “se = True“. As we see in the examples, this creates a dark-grey region around the smooth line. This dark grey area indicates the confidence interval (0.95 by default).
# If we set “se = False“, it will remove the confidence interval.     lm is used to fit linear models
```
